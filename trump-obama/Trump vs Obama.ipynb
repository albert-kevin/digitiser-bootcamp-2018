## Approach

What makes a successful president ? Success is an arbitrary quality which depends on a number of factors. Taking inspiration from the thesis "What Makes A Successful President? Measuring Presidential Success Through Our Last Five Presidents" by Shirin Sharif which is publicly available from Stanford University website, and also to keep the scale of this project simple, I have decided upon a few key factors that determine a Presidents Success. 

The Factors that are of interest are:
1. National Debt
(DataSet Downloaded From https://fred.stlouisfed.org/series/GFDEBTN) (Quarterly)
2. Unemployability Rates 
(DataSet Downloaded From https://fred.stlouisfed.org/series/LRUN64TTUSM156S) (Monthly)
3. Stock Market Index (S&P 500)
(DataSet Downloaded From https://www.macrotrends.net/2488/sp500-10-year-daily-chart (Daily)



```python
#Importing necessary modules

import pandas as pd
import matplotlib.pyplot as plt

# Initializing variables

obama_debt_inital = 0
obama_debt_final = 0
trump_debt_initial = 0
trump_debt_final = 0
obama_debt_list = []
trump_debt_list = []
filename = 'national debt.csv'

# Defining the extract_debt function. It is made wot be compatible with all the datasets for this project

def extract_data(filename, year, column, skip = None): # The column and skip parameters are present to accomodate the third dataset
    '''Uses filename and year to return a list of records with the desired year'''
    debt_data = pd.read_csv(filename, skiprows = skip )
    required_dates = []
    for x in debt_data[column]:
        if int(x[0:4]) == year:
            required_dates.append(True)
        else:
            required_dates.append(False)
    return list(debt_data.iloc[required_dates, 1])

# Using extract_debt function to get the list of debt during the first year of tenure

obama_debt_list = extract_data(filename, 2009, 'DATE')           
trump_debt_list = extract_data(filename, 2017, 'DATE')

# Scaling the debt lists in order to visualize them both as though they started with zero debt 
# Inflation is not taken into account between the years as it does not change the overall trend all that much

obama_debt_initial = obama_debt_list[0]
obama_debt_final = obama_debt_list[-1]
trump_debt_initial = trump_debt_list[0]
trump_debt_final = trump_debt_list[-1]

for i in range(len(obama_debt_list)):
    obama_debt_list[i] -= obama_debt_initial
for i in range(len(trump_debt_list)):
    trump_debt_list[i] -= trump_debt_initial

# Plotting Graphs using maptplotlib.pyplot

axis = ['Q1','Q2','Q3','Q4']
plt.plot(axis, trump_debt_list)
plt.plot(axis, obama_debt_list)
plt.legend(['Trump','Obama'])
plt.xlabel('Year in Quarters Intervals')
plt.ylabel('USD in Millions')
plt.title('Total Debt Accrued In First Year of Presidency')
plt.show
```

```python
# Using extract_data to get unemployment list as the dataset formats are similiar

filename = 'unemployment rate.csv'
obama_unemploy_list = extract_data(filename, 2009, 'DATE')           
trump_unemploy_list = extract_data(filename, 2017, 'DATE')

# Plotting Graphs
axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.plot(axis,trump_unemploy_list)
plt.plot(axis,obama_unemploy_list)
plt.legend(['Trump','Obama'])
plt.xlabel('Months')
plt.ylabel('Unemployability %')
plt.title('Unemployability Rates During First Year Of Presidency')
plt.show
```

```python
# The paramaters are not the same as the other two as this dataset is from another site and formatted differently 
# Modifications to the extract_data function had to be made to accomodate all datasets used

filename = 'stock market.csv'

obama_index = extract_data(filename, 2009, 'date', 12)           
trump_index = extract_data(filename, 2017, 'date', 12)

#Scaling DataSets to see rise in index as though they both started at 0

obama_initial = obama_index[0]
obama_final = obama_index[-1]
trump_initial = trump_index[0]
trump_final = trump_index[-1]

for i in range(len(obama_index)):
    obama_index[i] -= obama_initial
for i in range(len(trump_index)):
    trump_index[i] -= trump_initial

# Plotting Graphs

plt.plot(trump_index)
plt.plot(obama_index)
plt.legend(['Trump','Obama'])
plt.xlabel('Days')
plt.ylabel('S&P 500 Index')
plt.title('Index Growth During First Year Of Presidency')
plt.show
```
